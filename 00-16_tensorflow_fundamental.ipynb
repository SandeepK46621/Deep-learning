## Lecture 24
**Matrix Multiplication**

```python
# matrix multiplication
A = tf.constant([[10, 7],
                 [7,10]])
B = tf.constant([[11,15],
                 [12,19]])

C = tf.matmul(A,B)
C
```

```python
# if we use an operator to do the multiplicaiton of tensor it will do it element vice

A*B==C
# not equal beacuse there is a element based multiplicaiton
```

```python
# matrix multipalicaion with @ symbol

A @ B
```

```python
# matrix multiplicaiton of two different shape tensor

A = tf.constant([[1,2],
                 [3,4],
                 [5,6]])
B = tf.constant([[4 ,5 ,6],
                 [7 ,8 ,9]])

tf.matmul(A,B)
```

## Lecture 25
**Matrix multiplicaiton rules**

1. inner dimention should match
2. resultin Matrix will be of outer dimention

```python
A = tf.constant([[1,2,3],
                [4,5,6],
                [7,8,9]])

B = tf.constant([[1,2,3],
                 [4,5,6]])

A.shape , B.shape
```

```python
# inorder to multiply  the we can reshape a tensor
A @ (tf.reshape(B,shape=[3,2]))
```

```python
# inorder to multiply  the we can transpose a tensor
A @ (tf.transpose(B))
```

## Lecture 26

```python
tf.matmul(A, tf.transpose(B))
```

## Lecture 28

**aggregate function in tensor**

1. absolute
2. mean
2. min
3. max
4. sum
5. standard daviation
6. variance

```python
# creating a tesor to apply all aggregate function

tensor_rank_1  = tf.constant(np.random.randint(low=0 ,high=25 ,size=50))
tensor_rank_1
```

```python
# gives absolute value by element
tensor_test = tf.constant([4,-2,-8])

tf.abs(tensor_test)
```

```python
print("maximum of tensor - ", tf.reduce_max(tensor_rank_1))
print("minimum of tensor - ", tf.reduce_min(tensor_rank_1))
print("mean of tensor - ", tf.reduce_mean(tensor_rank_1))
print("sum of tensor - ", tf.reduce_sum(tensor_rank_1))
tensor_rank_1 = tf.cast(tensor_rank_1, dtype= tf.float32)  # to get variance and standard daviation convert the tensor into flaot
print("variance of tensor - ",tf.math.reduce_variance(tensor_rank_1))
print("standard daviation of tensor - ",tf.math.reduce_std(tensor_rank_1))
```
