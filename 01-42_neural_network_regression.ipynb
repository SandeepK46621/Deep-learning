### **Creating a Model**

**1.Creating model** :- **1.1** defining input layer
                    **1.2** defining hidden layer
                    **1.3** defining output layer

**2.Comiling model** :- **2.1** defining loss function (tells the model how away from real value it is )
                    **2.2** defining optimzer (helps our model to improve the pattern learning)
                    **2.3** defining evaluation metrix (what we can use to interpret the perforamce of model)

**3.Fitting model** :- letting the model find the pattern between X and Y

````python
tf.random.set_seed(42)                                # to get a generative model

# creating a model
model = tf.keras.Sequential([                          # sequential is group of linear stacked layer to keras API
    tf.keras.layers.Dense(1)                           # dense is a layer with a given number of unit neurons
])

# compilinig model
model.compile(loss= tf.keras.losses.mae ,               # mae short form is mean absolute error    mean(abs(y_predict , y_true))
              optimizer= tf.keras.optimizers.SGD(),     # sgd fullform is stocastic gradeant desent
              metrics= ["mae"]                          # metrix evaluation mean absolute error
              )
# Fitting model
model.fit(tf.expand_dims(x,axis=-1),y,epochs=5)                                 # epochs is number of laps that can be runned

````

````python
model.predict([17.])
````

* Total parameter - are the number of parameter in the model
* Trainable parameter - are the number of parameter(partten) the model can update as it trains
* Non-trainable parameter - this paramter are not updated during training as we import pre trained model **transfer learning** we dont want a pre trained model to change

####Research
**Trained parameter**
Trained parameter between two layers is   
n= number of unit in layer 1
m= number of unit in layer 2

trainabe parameter =   no.of_unit_in_layer_1 * no.of_unit_of_layer_2  + biased tems

trainabe parameter =  (n*m) + m

simplified form =  (n+1)*m



**Total parameter**

The number of internal parameters in a neural network is total number of weights + the total number of biases. The total number of weights equals the sum of the products of each pair of adjacent layers. The total number of biases is equal to the number of hidden neurons + the number of output neurons.

For example, consider a neural network with 13 input neurons, two hidden layers of 5 and 4 neurons, and an output layer of 3 neurons. The total number of weights is ( 13 * 5 ) + ( 5 * 4 ) + ( 4 * 3 ) = 97. The total number of biases is 5 + 4 + 3 = 12. Which makes the total number of parameters 97 + 12 = 109.

````python
# Creating a funciton to plot the data training testing and predictiona


def plot_prediction(train_data=x_train,
                    train_label=y_train,
                    test_data=x_test,
                    test_label=y_test,
                    test_pred=y_pred):
  # plotting the train data , test data aswell as predicted data
  plt.scatter(train_data , train_label , c="b" , label="Training Data")
  plt.scatter(test_data , test_label , c="g", label="Testing Data")
  plt.scatter(test_data, test_pred ,c ="r" ,label="Predicted Data")
  plt.legend()
````
